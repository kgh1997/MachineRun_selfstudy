```python
'''
보스턴 집값 데이터를 활용하여 다음 두 가지 과제를 수행하고 각각의  score 를 확인하세요.( 폴드 수 : 5, shuffle=False)

1. KFold 객체를 활용하되  인덱스 집합을 리턴받아 학습을 수행하여 r2 값을 구하시오 ( 소숫점 이하 3자리)
2. cross_val_score() 를 활용하여 r2값을 구하시오(소숫점 이하 세자리)

R2: [ 0.639 0.714 0.587 0.079 -0.253] 
average R2: 0.353
'''
```

```python
import numpy as np
import pandas as pd
from sklearn import datasets
boston = datasets.fetch_openml('boston')
boston
```

```python
boston.data.info()
```

```python
boston.target_names
```

```python
# 데이터 준비
X = boston.data.values
y = boston.target
X.shape, y.shape
```

```python
from sklearn.model_selection import KFold
kfold = KFold()
list(kfold.split(X))
```

```python
r2_scores = []
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
for train_idx, test_idx in kfold.split(X):
    X_train = X[train_idx]
    X_test = X[test_idx]
    y_train = y[train_idx]
    y_test = y[test_idx]
    Lr.fit(X_train, y_train)
    y_pred = Lr.predict(X_test)
    r2 = r2_score(y_test,y_pred)
    r2_scores.append(r2)
#r2_scores # 이 다섯개의 오차의 평균이 예측 오차
for i,r2 in enumerate(r2_scores):
    print(i+1, f' - R2 = {r2:.3f}')
print(f' average R2 = {np.mean(r2_scores)}')
```

```python
# cross_val
from sklearn.model_selection import cross_val_score
c_v = cross_val_score(Lr,X=X, y=y, cv=5)
np.set_printoptions(precision=3)
print(c_v, np.mean(c_v))
```
